## Approach 2

```python
print(f'this is x{x}')
print(f'this is y{y}')
```

```python
xtrain=x[:-5]
ytrain=y[:-5]


xtest=x[17:]
ytest=y[17:]

print(xtrain)
print(ytrain)
print(xtest)
print(ytest)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# perceptron model

model=tf.keras.Sequential([
                tf.keras.layers.Dense(units=1,input_shape=[1])
])
```

```python
model.compile(optimizer='adam',loss='mae')
```

```python
model.fit(x=xtrain , y=ytrain , validation_data=(xtest,ytest) , epochs=500)
```
